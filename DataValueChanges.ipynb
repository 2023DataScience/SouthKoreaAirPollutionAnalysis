```python
#### data value changes ####
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import MinMaxScaler
df = pd.read_csv('south-korean-pollution-data.csv')
print(df)
```

```python
#### 1.Cleaning dirty data

## missing data가 있는 행 제거
# 우리가 임의로 입력해야하니까 replace보단 제거로 가는 게 좋을듯!
df.dropna(inplace=True)
df_clean = df.copy()
print(df_clean)
```

```python
##wrong data가 있는 행 제거

# feature 별로 data type 확인
data_types = df_clean.dtypes

# type과 맞지 않은 값을 가지고 있는 행 제거
for column in df_clean.columns:
    if data_types[column] == 'int64':
        df_clean = df_clean[df_clean[column].apply(lambda x: isinstance(x, int))]
    elif data_types[column] == 'float64':
        df_clean = df_clean[df_clean[column].apply(lambda x: isinstance(x, float))]
    elif data_types[column] == 'object':
        df_clean = df_clean[df_clean[column].apply(lambda x: isinstance(x, str))]

print(df_clean)

# unusable data를 지우는 과정의 경우 data reduction이랑 같기 때문에 제외하고 작성함
```

```python
##outlier가 있는 행 제거

pollution_features = ['pm25', 'pm10', 'o3', 'no2', 'so2', 'co']

# outlier 제거 함수 정의
def remove_outliers(df_clean, feature):
    # Z-score 계산
    z_scores = (df_clean[feature] - df_clean[feature].mean()) / df_clean[feature].std()
    # Z-score의 절대값이 3보다 큰 행 선택
    outlier_indices = np.abs(z_scores) > 3
    # 이상치를 가지고 있는 행 제외
    df_clean = df_clean[~outlier_indices]
    return df_clean

# 각 feature에 대해 outlier 제거
for feature in pollution_features:
    df_clean = remove_outliers(df_clean, feature)

print(df_clean)
```

```python
#### 3.Data scaling ( Min-max scaling )

# Min-Max Scaling 객체 생성
scaler = MinMaxScaler()
# 각 feature에 대해 Min-Max Scaling 수행
df_clean[pollution_features] = scaler.fit_transform(df_clean[pollution_features])

# 수정된 데이터프레임 출력 (선택 사항)
print(df_clean)
```
