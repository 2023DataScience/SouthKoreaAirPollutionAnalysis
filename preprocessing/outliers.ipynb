preprocessing: Detecting Outliers

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터셋을 DataFrame으로 로드
df = pd.read_csv('data/semifinaldata_4.csv')

# 이상치를 찾을 변수들 선택
columns_to_check = ['pm25', 'pm10', 'so2', 'o3', 'no2', 'co', 'AVG']

# 이상치를 확인하는 함수 정의
def find_outliers_iqr(df, column):
    # 1사분위수(Q1)와 3사분위수(Q3) 계산
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    
    # IQR 계산
    IQR = Q3 - Q1
    
    # 이상치 경계값 계산
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # 이상치 확인
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    
    return outliers

# 이상치를 시각화하는 함수 정의
def plot_outliers_boxplot(df, column):
    plt.figure(figsize=(10, 6))
    plt.boxplot(df[column])
    plt.title(f"Box Plot for {column}")
    plt.ylabel(column)
    plt.show()

# 각 변수별로 이상치 확인 및 시각화
for column in columns_to_check:
    outliers = find_outliers_iqr(df, column)
    print(f"Outliers for {column}:")
    print(outliers)
    print()
    plot_outliers_boxplot(df, column)
```
