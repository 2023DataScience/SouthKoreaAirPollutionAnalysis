preprocessing: Handling Missing Value


```python
####### preprocessing ( pollution data )#######
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler


# Read the pollution data from a CSV file
df = pd.read_csv('data/pollution_data.csv', encoding='utf-8')


pollutants = ['pm25', 'pm10', 'o3', 'no2', 'so2', 'co']


# Drop rows with NaN values
df = df.dropna()


# Drop rows with string values in pollutants columns
for pollutant in pollutants:
    df = df.drop(df[df[pollutant] == '-'].index, axis=0)


# Remove asterisks from pollutant values
for pollutant in pollutants:
    df[pollutant] = df[pollutant].str.replace("*", "")
    df[pollutant] = df[pollutant].str.replace("**", "")


# Convert the pollutant columns to float
df[pollutants] = df[pollutants].astype(float)


# Function to find outliers using IQR
def find_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers


# Function to plot outliers using boxplot
def plot_outliers_boxplot(df, column):
    plt.figure(figsize=(10, 6))
    plt.boxplot(df[column])
    plt.title(f"Box Plot for {column}")
    plt.ylabel(column)
    plt.show()


# Find and visualize outliers for each pollutant
for pollutant in pollutants:
    outliers = find_outliers_iqr(df, pollutant)
    print(f"Outliers for {pollutant}:")
    print(outliers)
    print()
    plot_outliers_boxplot(df, pollutant)


# Data normalization
scaler = StandardScaler()
data_std = scaler.fit_transform(df[pollutants])
data_std = pd.DataFrame(data_std, columns=pollutants)
data_std['City'] = df['City']
data_std['District'] = df['District']


print(data_std)
```

Preprocessing: Merging data(initial version)

```python
####### Preprocessing ( pollution )#######
####### merge pollution data and life expectancy data #######
 
import pandas as pd
import numpy as np


# Read the life expectancy data
try:
    df = pd.read_csv('data/life_avg.csv', encoding='utf-8')
except UnicodeDecodeError:
    df = pd.read_csv('data/life_avg.csv', encoding='cp949')


# Keep data for years 2010-2020
df = df.rename(columns={'지표연도': 'year', '시군구': 'City', '시도': 'District', '평균 기대수명': 'AVG'})
df = df[df['성별'] == '전체']
df = df[['City', 'year', 'District', 'AVG']]
df = df[df['City'] != '전체']
df = df[(df['year'] >= 2010) & (df['year'] <= 2020)]


# Read the pollution data
df_pollute = pd.read_csv('data/pollution_final.csv')
df_pollute['year'] = df_pollute['date'].astype(str).str[:4]
df_pollute.drop('date', axis=1, inplace=True)
df_pollute['year'] = df_pollute['year'].astype(int)


# Remove suffixes from city names
excluded_cities = ['부산광역시', '대구광역시', '인천광역시', '광주광역시', '대전광역시', '울산광역시', '서울특별시', '세종특별자치시']


def remove_suffix(city):
    if city in excluded_cities:
        return city
    else:
        idx = city.rfind('시')
        if idx != -1:
            return city[:idx]
        else:
            return city


df['City'] = df['City'].apply(remove_suffix)


# Merge life expectancy and pollution data
merged_df = df_pollute.merge(df[['City', 'year', 'AVG']], on=['City', 'year'], how='inner')


# Remove rows with missing values in AVG column
merged_df = merged_df.dropna(subset=['AVG'])


# Rearrange columns and reset index
merged_df = merged_df[['District', 'City', 'year', 'pm25', 'pm10', 'so2', 'o3', 'no2', 'co', 'AVG']]
merged_df = merged_df.reset_index(drop=True)


print(merged_df)
```
